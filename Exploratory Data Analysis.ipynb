## Visualizing Numerical Variables

```python
sns.histplot(cardata['Kms_Driven'])
```

```python
sns.displot(cardata['Kms_Driven'], kind= 'kde') # method 1
```

```python
sns.displot(data= cardata, x = 'Kms_Driven', kind = 'kde') # method 2
```
